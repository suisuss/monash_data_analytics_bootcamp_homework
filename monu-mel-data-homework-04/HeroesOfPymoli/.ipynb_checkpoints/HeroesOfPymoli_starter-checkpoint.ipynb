```python
# Dependencies and Setup
import pandas as pd

# File to Load
file_to_load = "Resources/purchase_data.csv"

# Read purchase_data
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Can do any of these 3
total_purchases = purchase_data['Purchase ID'].count()
total_purchases = purchase_data.shape[0]
total_purchases = len(purchase_data.index) # This is fastest
total_purchases
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Get explicit data
price_series = purchase_data['Price']
item_name_series= purchase_data['Item Name']

# Calculations
purchasing_analysis_total_revenue = price_series.sum()
purchasing_analysis_average_price = round(price_series.mean(), 2)
purchasing_analysis_num_of_unique = item_name_series.nunique()

# Create DF
data = {
    'Total Revenue': [purchasing_analysis_total_revenue], 
    'Average Price': [purchasing_analysis_average_price], 
    'Number of Unique Items': [purchasing_analysis_num_of_unique], 
    'Total Purchases': [total_purchases]}
purchasing_analysis_summary = pd.DataFrame.from_dict(data)
purchasing_analysis_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Get Gender groupby
purchase_data_gender_groupby = purchase_data.groupby(by=['Gender'])

# Calculcations
gender_demographics_count = purchase_data_gender_groupby.size()
gender_demographics_percentage = purchase_data_gender_groupby.size() / purchase_data_gender_groupby.size().sum() * 100

# Create DF
data = {'Purchase Count': gender_demographics_count, 'Percentage': gender_demographics_percentage}
gender_demographics_df = pd.DataFrame.from_dict(data)
gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculations on purchase data grouped by gender | .agg method is the best way to go with group by objects
# .agg returns DF so used squeeze method to convert into series (easier for other calculations like average_total_purchase_per_person)
purchasing_analysis_genders_count = purchase_data_gender_groupby.agg({"SN" : "nunique"}).squeeze() # Get gender count in terms of unique SNs
purchasing_analysis_genders_purchase_count = purchase_data_gender_groupby.agg({"Purchase ID" : "count"}).squeeze() # Get purchase count by counting Purchase ID column
purchasing_analysis_genders_average_price = round(purchase_data_gender_groupby.agg({"Price" : "mean"}), 2).squeeze() # Get mean of values in price column
purchasing_analysis_genders_purchase_total = purchase_data_gender_groupby.agg({"Price" : "sum"}).squeeze() # Get sum of values in price column
purchasing_analysis_genders_average_total_purchase_per_person = round(purchasing_analysis_genders_purchase_total / purchasing_analysis_genders_count, 2)

# Two methods of creating new DF:
# 1. Build off of purchase_data_genders_basic
purchasing_analysis_genders_df = gender_demographics_df.copy()
purchasing_analysis_genders_df['Gender Count'] = purchasing_analysis_genders_count
purchasing_analysis_genders_df['Purchase Count'] = purchasing_analysis_genders_purchase_count
purchasing_analysis_genders_df['Average Price'] = purchasing_analysis_genders_average_price
purchasing_analysis_genders_df['Purchase Total'] = purchasing_analysis_genders_purchase_total
purchasing_analysis_genders_df['Avg Total Purchase per Person'] = purchasing_analysis_genders_average_total_purchase_per_person
del purchasing_analysis_genders_df['Percentage']

# 2. Concatenated values as new DF
purchasing_analysis_genders_df = pd.concat(
    [
        purchasing_analysis_genders_count,
        purchasing_analysis_genders_purchase_count,
        purchasing_analysis_genders_average_price,
        purchasing_analysis_genders_purchase_total,
        purchasing_analysis_genders_average_total_purchase_per_person
    ], axis=1)
purchasing_analysis_genders_df.columns = ['Gender Count', 'Purchase Count', 'Average Price', 'Purchase Total', 'Avg Total Purchase per Person']
purchasing_analysis_genders_df
```

```python
## Age Demographics
```

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create bins
age_group_bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
# Create labels
age_group_labels = ["< 10","10 - 14", "15 - 19", "20 - 24", "25 - 29", "30 - 34", "35 - 39", ">= 40"]
# Add to orginal DF
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_group_bins, labels=age_group_labels)

# Get Age Group groupby 
purchase_data_age_groupby = purchase_data.groupby('Age Group')

# Calculations
age_demographics_count = purchase_data_age_groupby.size()
age_demographics_percentage = round(purchase_data_age_groupby.size() / purchase_data_age_groupby.size().sum() * 100, 2)

# Create DF
data = {'Count': age_demographics_count, 'Percentage': age_demographics_percentage}
age_demographics_df = pd.DataFrame(data)
age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculations
purchasing_analysis_age_purchase_count = purchase_data_age_groupby.agg({"Purchase ID" : "count"}).squeeze()
purchasing_analysis_age_average_price = round(purchase_data_age_groupby.agg({"Price" : "mean"}), 2).squeeze()
purchasing_analysis_age_purchase_total = purchase_data_age_groupby.agg({"Price" : "sum"}).squeeze()
purchasing_analysis_age_average_total_purchase_per_person = round(purchasing_analysis_age_purchase_total / purchasing_analysis_age_purchase_count, 2)

# Create DF
purchasing_analysis_age_df = pd.concat([purchasing_analysis_age_purchase_count, purchasing_analysis_age_average_price, purchasing_analysis_age_purchase_total, purchasing_analysis_age_average_total_purchase_per_person], axis=1)
purchasing_analysis_age_df.columns = ['Purchase Count', 'Average Price', 'Purchase Total', 'Avg Total Purchase per Person']
purchasing_analysis_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get SN groupby
purchase_data_sn_groupby = purchase_data.groupby('SN')

# Calculations
top_spenders_purchase_count = purchase_data_sn_groupby.agg({"Purchase ID" : "count"}).squeeze() # Get purchase count by counting Purchase ID column
top_spenders_average_price = round(purchase_data_sn_groupby.agg({"Price" : "mean"}), 2).squeeze() # Get mean of values in price column
top_spenders_purchase_total = purchase_data_sn_groupby.agg({"Price" : "sum"}).squeeze()

# Create DF
top_spenders_df = pd.concat([top_spenders_purchase_count, top_spenders_average_price, top_spenders_purchase_total], axis=1)
top_spenders_df.columns = ['Purchase Count', 'Average Price', 'Purchase Total']
top_spenders_df = top_spenders_df.sort_values(by='Purchase Total', ascending=False).head()
top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Get Item groupby
purchase_data_item_groupby = purchase_data.groupby(['Item ID', 'Item Name'])

# Calculations
most_popular_items_count = purchase_data_item_groupby.agg({"Purchase ID" : "count"}).squeeze()
most_popular_items_price = round(purchase_data_item_groupby['Price'].mean(), 2)
# ^^^ Most prices are consistant but a few may not be (Crucifer for example has inconsistant prices [1.99, 3.26, 3.26, 3.26]),
# so I decided to calculate average price ^^^
most_popular_items_total = purchase_data_item_groupby.agg({"Price" : "sum"}).squeeze()

# Create DF
most_popular_items_df = pd.concat([most_popular_items_count, most_popular_items_price, most_popular_items_total], axis=1)
most_popular_items_df.columns = ['Purchase Count', 'Average Price', 'Purchase Total']
most_popular_items_df.sort_values(by='Purchase Count', ascending=False).head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable_items_df = most_popular_items_df.sort_values(by='Purchase Total', ascending=False).head()
most_profitable_items_df
```
