## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import random

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = mouse_metadata.merge(study_results, on='Mouse ID', how='outer')

# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
total_mice = len(combined_df.index)
total_unique_mice = combined_df['Mouse ID'].nunique()
print(f"Total mice: {total_mice}")
print(f"Total unique mice: {total_unique_mice}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_df[combined_df.duplicated(subset=('Mouse ID', 'Timepoint')) == True]
duplicate_mice_ID = duplicate_mice['Mouse ID']
duplicate_mice_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_df = combined_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'])
```

```python
# Checking the number of mice in the clean DataFrame.
total_mice = len(clean_combined_df.index)
total_unique_mice = clean_combined_df['Mouse ID'].nunique()
print(f"Total mice: {total_mice}")
print(f"Total unique mice: {total_unique_mice}")
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
clean_combined_regimen_groupby = clean_combined_df.groupby('Drug Regimen')
summary_stats_regimen = clean_combined_regimen_groupby.agg({'Tumor Volume (mm3)': [np.mean, np.median, np.var, np.std, st.sem]})
summary_stats_regimen
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

regimen_mouseID_groupby = clean_combined_df.groupby(['Drug Regimen', 'Mouse ID'])

final_tumor_volume_df = regimen_mouseID_groupby.last()['Tumor Volume (mm3)'].to_frame().reset_index()
final_tumor_volume_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ['Capomulin', 'Ramicane', 'Infubinol','Ceftamin']

# Calculate the IQR and quantitatively determine if there are any potential outliers.

final_tumor_volume_df_compact = final_tumor_volume_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(list).to_frame().reindex(regimens)
# ^^^ It was attracitve to go on with but doing .quantile([.25,.5,.75]) on a series of lists proved difficult. Ended up just using it for plots.

quartiles = final_tumor_volume_df.groupby('Drug Regimen').quantile([.25,.5,.75])
regimens_quartiles = quartiles.take([0, 1, 2, 3, 4, 5, 6, 7, 8, 21, 22, 23], axis=0)

capomulin = regimens_quartiles.iloc[:3, :]
ceftamin = regimens_quartiles.iloc[3:6, :]
infubinol = regimens_quartiles.iloc[6:9, :]
ramicane = regimens_quartiles.iloc[6:9, :]

regimens_quartiles_dict = dict(zip(regimens, [capomulin, ceftamin, infubinol, ramicane]))
regimens_outlier_list = list()
for regimen, regimens_quartile_df in regimens_quartiles_dict.items():
    print(regimen.upper())
    print('-' * len(regimen))
    lowerq = regimens_quartile_df['Tumor Volume (mm3)'][0]
    middleq = regimens_quartile_df['Tumor Volume (mm3)'][1]
    upperq = regimens_quartile_df['Tumor Volume (mm3)'][2]
    iqr = upperq-lowerq

    print(f"The lower quartile of Tumor Volume (mm3) is: {lowerq}")
    print(f"The upper quartile of Tumor Volume (mm3) is: {upperq}")
    print(f"The interquartile range of Tumor Volume (mm3) is: {iqr}")
    print(f"The the median of Tumor Volume (mm3) is: {middleq} ")

    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.\n")
    
    outliers = final_tumor_volume_df.loc[(final_tumor_volume_df['Drug Regimen'] == regimen) & ((final_tumor_volume_df['Tumor Volume (mm3)'] < lower_bound) | (final_tumor_volume_df['Tumor Volume (mm3)'] > upper_bound))]
    print(f"Here are any potential outliers: \n{outliers}\n")
    
    for mouse_id in outliers['Mouse ID']:
        regimens_outlier_list.append({regimen: mouse_id})

# #CLEAN
regimens_outlier_list
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.figure(figsize=(8,6))
plt.boxplot(final_tumor_volume_df_compact['Tumor Volume (mm3)'], labels=regimens)
plt.ylim(10, 80)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
time_vs_tumor_df = clean_combined_df[clean_combined_df["Mouse ID"] == "b128"][["Timepoint", "Tumor Volume (mm3)"]]

line_plot_df = time_vs_tumor_df[["Timepoint", "Tumor Volume (mm3)"]]

lines = line_plot_df.plot.line(title="Time vs Tumor (Mouse: b128)", figsize=(8,6))
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

# Getting data
weight_vs_tumor_df = clean_combined_df[clean_combined_df["Drug Regimen"] == "Capomulin"][["Weight (g)", "Tumor Volume (mm3)"]]

# Sorting data for Weight (g)
weight_vs_tumor_df_weight_sorted = weight_vs_tumor_df.sort_values(["Weight (g)"], ascending=True)

# Grouping by weight and getting average tumor volumes for each weight. Reseting index for plotting
weight_vs_tumor_df_weight_sorted_mean = weight_vs_tumor_df_weight_sorted.groupby("Weight (g)")["Tumor Volume (mm3)"].mean().reset_index()

capomulin_scatter = weight_vs_tumor_df_weight_sorted_mean.plot(kind='scatter', x='Weight (g)', y='Tumor Volume (mm3)', grid=True, figsize=(8,6))
#capomulin_scatter
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Getting plot values
x_values = weight_vs_tumor_df_weight_sorted_mean["Weight (g)"]
y_values = weight_vs_tumor_df_weight_sorted_mean["Tumor Volume (mm3)"]

# Perfomring linear regression calc and getting values
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)

# Getting regression values for plotting
regression_values = x_values * slope + intercept

# Defining equation
line_equation = f"y = {round(slope,2)}x + {round(intercept,2)}"

plt.figure(figsize=(8,6))
plt.scatter(x_values, y_values)
plt.plot(x_values, regression_values, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Weight Vs. Avg Tumor Vol ({line_equation})")
plt.show()
```
